## Data Visualization

### Matplotlib

```python
import matplotlib.pyplot as plt
```

```python
#from matplotlib import pyplot as plt
```

```python
#line plot

plt.plot([1,2,3],[4,5,2])
plt.show()
```

```python
plt.plot([1,2,3],[4,5,2])
plt.title("newgraph")
plt.xlabel("x-axis"),
plt.ylabel("y-label")
plt.show()
```

```python
#simple line graph
x1=[1,2,3]
y1=[4,5,2]
x2=[6,7,8]
y2=[9,10,7]
plt.plot(x1,y1,label='first',color='r',linewidth=5)
plt.plot(x2,y2,label='second',color='m',linewidth=5)
plt.title('style graph')
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.grid(color='blue')
plt.legend()
plt.show()
```

```python
#bar graph for categorical data
x1=[1,2,3]
y1=[4,5,2]
x2=[6,7,8]
y2=[9,10,7]
plt.bar(x1,y1,label="first")
plt.bar(x2,y2,label="second")
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.legend()
plt.show()
```

```python
# for continous values -histogram
sales_values=[12,10,5,7,0,14,15,24,10,11,15,30,18,11,10,18,30]
bins=[0,5,10,15,20,25,30]
plt.hist(sales_values,bins,histtype='bar',rwidth=0.6)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("histogram")
plt.show()
```

```python
#scatter graph
#relationship between x and y
x=[1,3,4,8,5,6,7]
y=[3,5,6,7,8,9,3]
plt.scatter(x,y,label="scatter",color='r')
plt.xlabel("x-axis")
plt.ylabel("y-label")
plt.title("scatter plot")
plt.legend()
plt.show()
```

```python
mylabels='python','c++','ruby','java'
sizes=[215,130,245,210]

mycolors=['gold','yellowgreen','lightcoral','lightblue']
plt.pie(sizes,labels=mylabels,colors=mycolors,autopct="%1.2f%%")
plt.show()
```

### seaborn

```python
#import seaborn
import seaborn as sb
```

```python
#inbuilt dataset in seaborn
sb.get_dataset_names()
```

```python
df=sb.load_dataset('iris')
sb.distplot(df['petal_length'])
plt.show()
```

```python
df=sb.load_dataset('iris')
sb.distplot(df['petal_length'],hist=False)
plt.show()
```

```python
df=sb.load_dataset('iris')
sb.distplot(df['petal_length'],kde=False)
plt.show()
```

```python
df=sb.load_dataset('iris')
sb.distplot(df['petal_length'])
sb.set_style("darkgrid")
plt.show()
```

```python
#set palette color
df=sb.load_dataset('iris')
sb.distplot(df['petal_length'])
sb.set_style("darkgrid")
sb.set_palette("autumn")
plt.show()
```

```python
#categorical data with continous data
df=sb.load_dataset('iris')
sb.boxplot(data=df,x='species',y='petal_length')
sb.set_palette("spring")
plt.show()
```

```python
df=sb.load_dataset('iris')
sb.swarmplot(data=df,x='species',y='petal_length')
sb.set_palette("spring")
plt.show()
```

```python
df=sb.load_dataset('iris')
sb.violinplot(data=df,x='species',y='petal_length')
sb.set_palette("spring")
plt.show()
```

```python
sb.pairplot(df,hue='species',diag_kind='kde',kind="scatter",palette="husl")
plt.show()
```
